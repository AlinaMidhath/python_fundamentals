```python
# prompt: Can you find basic stats information on our dataframe df, calculate measurements such as mean, median, mode, variancee, std, range, skewness, kurtosis

import pandas as pd
df = pd.read_csv('/content/Salaries.csv')

# Calculate mean, median, mode, variance, standard deviation, range, skewness, and kurtosis for 'salary'
print("Mean:", df['salary'].mean())
print("Median:", df['salary'].median())
print("Mode:", df['salary'].mode()[0])  # Access the first mode if multiple exist
print("Variance:", df['salary'].var())
print("Standard Deviation:", df['salary'].std())
print("Range:", df['salary'].max() - df['salary'].min())
print("Skewness:", df['salary'].skew())
print("Kurtosis:", df['salary'].kurt())
```
